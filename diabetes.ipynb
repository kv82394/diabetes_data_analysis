```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# loading the datasets
train_data = np.loadtxt('training_diabetes_x_y_train.csv', delimiter = ',')
test_data = np.loadtxt('test_diabetes_x_test.csv', delimiter = ',')
```

```python
# creating numpy array for ready to parameter
arr = np.arange(0,10)
x_train = train_data[:,arr]
y_train = train_data[:,10]
print(x_train.shape,y_train.shape)
```

```python
# spliting trining data for finding the score
from sklearn import model_selection
x_train , x_test , y_train ,y_test = model_selection.train_test_split(x_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
```

```python
print(x_train.shape,y_train.shape)
```

```python
# training model 
algo.fit(x_train,y_train)
```

```python
# predicting output for testing data on given training data
y_test_predict_on_train_data = algo.predict(x_test)
print(y_test_predict_on_train_data)
```

```python
# output for training data 
y_train_predict_on_train_data = algo.predict(x_train)
print(y_train_predict_on_train_data)
```

```python
# score on training and testing data
print("Score on testing data : ",algo.score(x_test, y_test))
print("Score on training data : ",algo.score(x_train, y_train))
```

```python
# predicting output for testing data
y_test_predict = algo.predict(test_data)
y_test_predict
```

```python
#converting into csv file
np.savetxt('test_diabetes_y_test.csv', y_test_predict,fmt = "%0.5f",delimiter=',')
```
